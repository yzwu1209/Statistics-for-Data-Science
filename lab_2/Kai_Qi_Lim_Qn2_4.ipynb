### Based on your EDA, select an appropriate hypothesis test.  (5 points)
Explain why your test is the most appropriate choice, focusing on its statistical assumptions.

* Considering that there are 1692 observations, the sample size is large enough for CLT to apply, so t-tests (and z-tests) are valid. However, due to some data skewness, a t-test is recommended. The t-test is robust to moderate deviations from normality, so you could argue that it is still valid here.
* Assuming that each participant provided their rating independently, the house18p variable is taken to be i.i.d. Age of participant is i.i.d. **And** *house18p* is independent of age, hence the two variables are considered unpaired.
* With the above observations, a parametric unpaired two-tailed t-test is recommended.
* Null hypothesis: The mean age of Republican voters is the same as Democrat voters.
* Alternative hypothesis: The mean age of Republican voters is not the same as Democrat voters.
* We decide to run two-tailed instead of one-tailed t-test in that we do not have strong arguments to , even though we are interested in answering whether one is higher than the other.
* The null hypothesis will be rejected if the test statistic falls in the rejection region for a p-value of 0.05.

### Conduct your test. (5 points)
Explain (1) the statistical significance of your result, and (2) the practical significance of your result.  Make sure you relate your findings to the original research question.

```r
# Run unpaired two-tailed t-test
t.test(democ$age, repub$age)

# Calculate the effect size
cohen.d(democ$age, repub$age)
```

**Conclusion**
* The p-value of $2.454e-07$ from the test showed statistical significance and suggests that we should reject the null hypothesis.
* The test lends support towards the alternative hypothesis that there **is a difference** between the age od Democrat and Republican votes.
* Although the sample size for the test is not small, it is unknown if there is a clustering effect, and it is not representative of the US population, so it is a stretch to state that there is a difference in general.

* Cohen's d estimate is -0.25, this is telling us that our means are about .25 standard deviations apart, which is considered small.
